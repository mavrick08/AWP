```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df
```

```python
plt.hist(df.height,bins=20,rwidth=0.7)
plt.xlabel('Height in inches')
plt.ylabel('Count')
plt.show()
```

```python
df['height'].mean()   #average value for column height.
```

```python
df['height'].std()
```

the average value of height column is **66.36** & the standard deviation value is **3.84**

## finding the upper limit

```python
upper_limit = df['height'].mean() + 3*df['height'].std()
```

```python
upper_limit
```

```python
lower_limit = df['height'].mean() - 3*df['height'].std()
```

```python
lower_limit
```

```python
df[(df['height']>upper_limit) | (df['height'] < lower_limit)]
```

```python
new_df = df[(df['height']<upper_limit) & (df['height'] > lower_limit)]
```

```python
new_df
```

```python
df
```

## z - score

```python
df['zscore'] = (df['height'] - df['height'].mean()) / df['height'].std()
```

```python
df
```

```python
df[df['zscore'] > 3]
```

```python
df[df['zscore'] < -3]
```

```python
df[(df['zscore'] < -3) | (df['zscore'] > 3)]
```

```python
df_nol_z_score = df[(df['zscore'] > -3) & (df['zscore'] < 3)]
```

```python
df_nol_z_score
```
