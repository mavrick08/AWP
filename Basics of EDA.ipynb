## Outlier detection / removal

```python
df_new = pd.read_csv('heights.csv')
```

```python
df_new
```

```python
df_new.describe()
```

```python
q1 = df_new['height'].quantile(0.25)
```

```python
q1
```

```python
q3 = df_new['height'].quantile(0.75)
```

```python
q3
```

```python
iqr = q3-q1
```

```python
iqr
```

```python
lower_limit = q1 - 1.5*iqr
```

```python
lower_limit
```

```python
upper_limit = q3+1.5*iqr
```

```python
upper_limit
```

```python
df_new1 = df_new.drop('name',axis=1)
```

```python
df_new1.info()
```

```python
df_new[(df_new1['height'] < lower_limit) | (df_new1['height'] > upper_limit)]
```

```python
df_new
```

```python
df_no_outlier = df_new[(df_new1['height'] > lower_limit) & (df_new1['height'] < upper_limit)]
```

```python
df_no_outlier
```

```python
sample_df
```

```python
sample_height = sample_df['height']
```

```python
sample_height
```

```python
sample_height.describe()
```

```python
q1 = sample_height.quantile(0.25)
```

```python
q1
```

```python
q3 = sample_height.quantile(0.75)
```

```python
q3
```

```python
iqr = q3-q1
```

```python
iqr
```

```python
lower_limit_ht = q1-1.5*iqr
upper_limit_ht = q3+1.5*iqr
```

```python
lower_limit_ht
```

```python
upper_limit_ht
```

```python
outlier_ht = sample_df[(sample_height < lower_limit_ht) | (sample_height > upper_limit_ht)]
```

```python
outlier_ht
```

```python
len(outlier_ht)
```

```python
no_outlier_ht = sample_df[(sample_height > lower_limit_ht) & (sample_height < upper_limit_ht)]
```
